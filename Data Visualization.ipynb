# Distribution of Fraudulent vs. Non-Fraudulent Transactions

```python
import plotly.express as px
 
# Count the number of fraud and non-fraud transactions
fraud_counts = df['isFraud'].value_counts()
 

# Plotting the pie chart
fig = px.pie(names=['Non-Fraud', 'Fraud'], values=fraud_counts.values, 
             title='Fraud vs Non-Fraud Transactions', hole=0.3)
fig.show()
```

# Distribution of Transaction Types Across Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
 

# Plotting
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=df)
plt.title('Distribution of Transaction Types')
plt.xlabel('Transaction Type')
plt.ylabel('Frequency')

plt.show()
```

# Comparison of Transaction Types: Fraudulent vs. Non-Fraudulent

```python
import seaborn as sns
import matplotlib.pyplot as plt

 

# Filter fraudulent and non-fraudulent transactions
fraudulent_transactions = df[df['isFraud'] == 1]
non_fraudulent_transactions = df[df['isFraud'] == 0]

# Plotting fraudulent transactions
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='type', data=fraudulent_transactions)
plt.title('Distribution of Transaction Types (Fraudulent)')
plt.xlabel('Transaction Type')
plt.ylabel('Frequency')

# Plotting non-fraudulent transactions
plt.subplot(1, 2, 2)
sns.countplot(x='type', data=non_fraudulent_transactions)
plt.title('Distribution of Transaction Types (Non-Fraudulent)')
plt.xlabel('Transaction Type')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
